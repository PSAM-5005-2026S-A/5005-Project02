```python
from ultralytics import YOLO
from yolo_utils import CustomizedTrainer, CustomizedValidator
```

```python
DATASET_PATH = "path/to/dataset"
```

```python
model = YOLO("yolo26n-cls.pt")
```

```python
model.train(data=DATASET_PATH, trainer=CustomizedTrainer, epochs=10, imgsz=224, batch=64)
```

```python
val_metrics = model.val(data=DATASET_PATH, validator=CustomizedValidator, imgsz=224, batch=64)
display(val_metrics)
val_metrics.top1
val_metrics.top5
```
